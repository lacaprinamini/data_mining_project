Questa linea di codice inizializza un oggetto KMeans, che  un algoritmo di clustering. L'obiettivo dell'algoritmo KMeans  suddividere un insieme di dati in 

k gruppi (cluster) distinti. Ecco una spiegazione dettagliata degli argomenti forniti:

n_clusters=3: Questo specifica che l'algoritmo dovrebbe cercare di dividere i dati in 3 cluster. Il valore di 

k  3 in questo caso.

n_init=10: L'algoritmo KMeans, nella sua forma base,  sensibile alle condizioni iniziali. Cio, i centroidi iniziali scelti possono influenzare il risultato finale. n_init specifica quante volte l'algoritmo sar eseguito con diversi centroidi iniziali. Il risultato finale sar il migliore output di queste 

n esecuzioni. In questo caso, l'algoritmo verr eseguito 10 volte.

max_iter=100: Questo  il numero massimo di iterazioni che l'algoritmo eseguir per una singola esecuzione. Se i centroidi si stabilizzano (cio, non cambiano tra le iterazioni) prima di raggiungere questo numero, l'algoritmo si fermer prima.

random_state=4: KMeans sceglie i centroidi iniziali in modo casuale (a meno che non vengano forniti esplicitamente). La specifica di un random_state assicura che la "casualit" sia riproducibile. Cio, ogni volta che esegui l'algoritmo con lo stesso random_state, otterrai gli stessi centroidi iniziali. Questo  utile per garantire risultati consistenti, specialmente durante la fase di test.

In sostanza, questa linea di codice inizializza un oggetto KMeans che cercher di dividere i dati in 3 cluster, eseguir l'algoritmo 10 volte con diversi centroidi iniziali per ogni esecuzione e utilizzer un massimo di 100 iterazioni per ogni esecuzione. La scelta dei centroidi iniziali sar basata su un seed casuale di 4.

### kth neighbor distance

```python
dist = pdist(X_minmax, 'euclidean')
dist = squareform(dist)
```

```python
sns.heatmap(dist, cmap="Greens_r", annot=False)
plt.show()
```

```python
k=4
kth_distances = []
for d in dist:  # d is a vector containing distances between the ith record and all the others
    index_kth_distance = np.argsort(d)[k]  # take the index of the kth nearest neighbor
    kth_distances.append(d[index_kth_distance])  # store the distance in a list
```

```python
plt.plot(range(0, len(kth_distances)), sorted(kth_distances))
plt.ylabel('distance from the {}th neighbor'.format(k))
         
plt.show()
```

### Hierarchical

```python
def get_linkage_matrix(model):
    # Create linkage matrix 
    
    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    return linkage_matrix

def plot_dendrogram(model, **kwargs):
    linkage_matrix = get_linkage_matrix(model)
    dendrogram(linkage_matrix, **kwargs)
```

```python
# setting distance_threshold=0 ensures we compute the full tree.
# it is the linkage distance threshold above which clusters will not be merged
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None, 
                                metric='euclidean', linkage='complete')
model = model.fit(X_minmax)
```

```python
plt.title("Hierarchical Clustering Dendrogram")
plot_dendrogram(model, truncate_mode='lastp', color_threshold=1.4)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```

```python
Z = get_linkage_matrix(model)
labels = fcluster(Z, t=1.4, criterion='distance')
```

```python
labels
```

```python
print('Silhouette', silhouette_score(X_minmax, labels))
```
